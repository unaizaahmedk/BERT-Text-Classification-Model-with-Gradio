```python
import torch
from transformers import BertForSequenceClassification, BertTokenizer

model_name = "bert-base-uncased"

model = BertForSequenceClassification.from_pretrained(model_name, num_labels=2)
tokenizer = BertTokenizer.from_pretrained(model_name)

model_save_path = "./bert_text_classification_model.pth"
torch.save(model.state_dict(), model_save_path)

tokenizer_save_path = "./bert_text_classification_tokenizer"
tokenizer.save_pretrained(tokenizer_save_path)

print(f"Model saved to {model_save_path}")
print(f"Tokenizer saved to {tokenizer_save_path}")
```

```python
import gradio as gr

model_save_path = "./bert_text_classification_model.pth"
tokenizer_save_path = "./bert_text_classification_tokenizer"

model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)
model.load_state_dict(torch.load(model_save_path))
tokenizer = BertTokenizer.from_pretrained(tokenizer_save_path)

def classify_text(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    logits = outputs.logits
    probabilities = torch.softmax(logits, dim=1)
    confidence, predicted_class = torch.max(probabilities, dim=1)
    labels = ["Negative", "Positive"]
    return {labels[i]: float(probabilities[0][i]) for i in range(len(labels))}

iface = gr.Interface(
    fn=classify_text,
    inputs=gr.Textbox(lines=2, placeholder="Enter text here..."),
    outputs=gr.Label(num_top_classes=2),
    title="BERT Text Classification",
    description="Classify text as Positive or Negative using a pre-trained BERT model.",
)
iface.launch()
```
